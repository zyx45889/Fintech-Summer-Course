```python
import pandas as pd
import numpy as np
import seaborn as sns
from palettable.cartocolors.sequential import TealGrn_2
```

```python
data = pd.read_csv("./train.csv")
```

```python
columns_info=pd.read_csv("data/feature_x.csv",index_col=0)
data=data.rename(columns={k:v for k, v in zip(columns_info["output_feat"],columns_info["description"])})
```

```python
import matplotlib.pyplot as plt
f, (ax1,ax2) = plt.subplots(figsize = (16, 8),nrows=2)
co=data.corr()
co["abs"]=abs(co["Y"])
new_co = co.sort_values(by=['abs'],ascending=False)
sns.heatmap(new_co,cmap=TealGrn_2.mpl_colormap,xticklabels=True,ax = ax1)
```
